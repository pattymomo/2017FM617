# Q1寫一個函式，抓取 Google Finance 的股價歷史資料

```python
import requests
```

```python
url= "http://finance.google.com/finance/historical"  # info from github datareader google daily file to find correct url
```

```python
params = {}
params['q'] = 'GE'
params['startdate'] = '2016-01-01'
params['ENDdate'] = '2017-10-31'
params['output'] = 'csv'
```

```python
r = requests.get(url, params=params)
```

```python
print(r.text)
```
